## 1.8 Broadcasting

As we all know, to perform an arithmatic operation, an array should be of same dimension. But Numpy overcomes this to some extent through Broadcasting.

Broadcasting is set of rules to perform arithmatic operations (e.g., addition, subtraction, multiplication, etc.)on arrays of different sizes.

Rules for broadcasting:

    Rule 1: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
    Rule 2: If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
    Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.


```python
#Example 1 for broadcasting

a = np.array([0, 1, 2])
b =  1

c = a+b

print(c)
```

```python
#Example 2 for broadcasting

a = np.ones((2, 3))
b = np.arange(3)

c = a+b

print(c)
```

Note: If rules of broadcasting doesn't follows, it throws an error

```python
# Below code throws an error as it doesn't follow the rules of broadcasting
a = np.ones((2, 3))
b = np.arange(3,1)

c = a+b

print(c)
```

```python
#Create Series object
s = pd.Series([3,4,5],['val1','val2','val3'])
k = pd.Series([1,2,3],['val1','val2','val3'])
  
#Create DataFrame df from two series
df = pd.DataFrame({'column1':s,'comun2':k})
```

**In the above method, we can change in one dataframe is reflected in the other dataframe as well.**

To overcome this we need to do deep copy of dataframe

```python
#Create Series object
s = pd.Series([3,4,5],['val1','val2','val3'])
k = pd.Series([1,2,3],['val1','val2','val3'])
  
#Create DataFrame df from two series
df = pd.DataFrame({'column1':s,'comun2':k})
```

```python
#print the values of data frame and its reference
print(df)

#perfom the deep copy of dataframe to another variable
Deep_variable_copy = df.copy(deep=True)

#Update the value of mass of earth in original datafram
df['column1']['val1']=8

#Print the dataframe
print(df)
print(Deep_variable_copy)
```

Codemat.JPG

A plot is a hierarchy of nested Python objects where a Figure object is the outermost container, which can contain Axes objects(holding actual plots). 


The Axes contains smaller objects such as tick marks, individual lines, legends, and text boxes. Almost every element of a chart including ticks and labels are maniputable Python object.

 There are two ways to plot a figure using matplotlib package i.e
 1. Stateful approach/functional approach

In stateful (state-machine) approach, plot is drawn using plt.plot() and other top-level pyplot functions by manipulating one Figure or Axes at a given time, without explicitly refering to it.
 
    syntax: plt.plot(xpoints, ypoints)

 2. Stateless approach/Object oriented approach
 
The object-oriented approach is more commonly used as it gives power to the user to create and manipulate multiple visualization i.e we can assign the figure and axes objects to variables and apply functions via methods on them.
  
    syntax: fig, ax = plt.subplots(nrows= , ncols= ) where it gives two objects figure and axes

## 3.4 Stateless approach/Object oriented approach

```python
# using subplots to create a Figure and Axes Object

days = list(range(1,9))
temp = np.array([19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1])

f, ax = plt.subplots() # returns tuple:
# f is the figure which can contain several plots/axes objects
#ax axes object

#setting the labels 
ax.set(xlabel='Day',
       ylabel='Temperature in celsius vs fahrenheit',
       title='sample temperature plot')

#styling the plot
ax.plot(days, temp,
        days, temp, "oy",
        days, (temp* 9/5 +32),
        days, (temp* 9/5 +32), "or")

plt.show()
```

```python
# Code taken from matplotlib org
# This figure shows the name of several matplotlib elements composing a figure

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter


np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xycoords='data',
            xytext=(3.3, 0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
```
