Understanding Deep copy and Shallow copy

```python
a= [["Rohit Sharma",45],["Virat Kohli",18], ["Ms dhoni", 7], ["Rishabh Pant", 17]]
```

```python
print(a)
print("\n Object type is", type(a))
print("\n a Address in the Memory", id(a))
```

```python
b = a #Assigning the Object "a" to "b"
```

```python
print(b)
print("\n Object b type is", type(b))
print("\n b Address in the Memory", id(b))
```

```python
if id(a)==id(b):
  print("a and b are referencing the same memory location")
```

You been told to keep only active players in the list, so instead of MS dhoni let's have KL Rahul in the list

```python
b[2] = ["KL Rahul", 1] #Making changes to Object reference "b"
```

```python
print(b)
print("\n a Address in the Memory", id(b))
```

#Notice the contents of the object reference "a" is also changed 

```python
print(a)
print("\n a Address in the Memory", id(a))
```

In  python, When you use assignment operator('=')to copy the objects, It actually creates a new variable that shares the reference of the original object. When you are working with mutable objects sometimes it may be necessary to create a cloan of an object so you can change real copy without modifying the original object.

Note: The difference between shallow copying and deep copying is only relevant for compound objects (objects that contain other objects, like lists or class instances)


In Python, there are two ways to create copies:

    Shallow Copy
    Deep Copy



A shallow copy constructs a new compound object and then (to the extent possible) inserts references into it to the objects found in the original.

A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.

#We use the copy module of Python for shallow and deep copy operations.

```python
import copy # import copy module for deep copy and shallow copy operations
```

#Performing Deep copy


```python
adeep= [["Rohit Sharma",45],["Virat Kohli",18], ["Ms dhoni", 7], ["Rishabh Pant", 17]]
```

```python
bdeep=copy.deepcopy(adeep) # performing Deepy copy 
```

```python
print(adeep)
print("\n Object type is", type(adeep))
print("\n a Address in the Memory", id(adeep))
```

```python
print(bdeep)
print("\n Object type is", type(bdeep))
print("\n a Address in the Memory", id(bdeep))
```

```python
if id(adeep)==id(bdeep):
  print("Both objects are referencing same memory locations")
else:
  print("They are referencing different memory locations")
```

```python
bdeep[2]=['KL Rahul',1]
```

```python
print(bdeep)
```

Notice modification to bdeep object doesn't reflect on adeep

```python
print(adeep)
```

#Performing Shallow copy

```python
ashallow = [["Rohit Sharma",45],["Virat Kohli",18], ["Ms dhoni", 7], ["Rishabh Pant", 17]]
```

```python
bshallow = copy.copy(ashallow) # performing shallow copy 
```

```python
print(ashallow)
print("\n Object type is", type(ashallow))
print("\n a Address in the Memory", id(ashallow))
```

```python
print(bshallow)
print("\n Object type is", type(bshallow))
print("\n a Address in the Memory", id(bshallow))
```

```python
if id(ashallow)==id(bshallow):
  print("Both objects are referencing same memory locations")
else:
  print("They are referencing different memory locations")
```

```python
bshallow[2][0]= "Javagal Srinath"
```

```python
print(bshallow)
```

Notice the changes in the child object of "b", is reflected in child object of "a"

```python
print(ashallow)
```
